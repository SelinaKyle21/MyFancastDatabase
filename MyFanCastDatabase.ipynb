## Casting Functions

```python
currentYear = 2024 # update to the calendar year of the movie production
```

```python
# A function for casting a character based on:
# - Age
# - Gender

def castCharacter(age, gender, ageRange = 5, dataFrame = df):
    birthYear = currentYear - age
    # print(birthYear)
    castingFilter = ((dataFrame['Birth_Year'] <= birthYear + ageRange) & (dataFrame['Active'] == 1) & (dataFrame['Birth_Year'] > birthYear - ageRange) & (dataFrame['Gender'] == gender))
    castingOptions_Info = dataFrame.loc[castingFilter, ['Name', 'Gender', 'Birth_Year','Nationality']]
    # print(castingOptions_Info.dtypes)
    return castingOptions_Info
```

```python
SleepingBeauty = castCharacter(16, 'F')
SleepingBeauty
```

```python
KingWilhelm = castCharacter(50, 'M', 10)
KingWilhelm
```

```python
# A function for casting a character based on:
# - Nationality

def castByNationality(nationality, dataFrame = df):
    castingFilter = ((dataFrame['Nationality'] == nationality))
    castingOptions_Info = dataFrame.loc[castingFilter, ['Name', 'Gender', 'Birth_Year','Nationality']]
    # print(castingOptions_Info.dtypes)
    return castingOptions_Info
```

```python
CaptainAmerica = castByNationality('American')
CaptainAmerica
```

```python
KingGeorge = castByNationality('English', castCharacter(40, 'M', 20))
KingGeorge
```

```python
# A function for choosing an actor out of a dataset
def chooseActor(actorName, dataFrame = df):
    choiceFilter = (dataFrame['Name'] == actorName)
    return dataFrame[choiceFilter]
```

```python
# A function for filtering a casted actor out of the dataset
def castActor(actorName, dataFrame = df):
    filterOutChoice = (dataFrame['Name'] != actorName)
    return dataFrame[filterOutChoice]
```
